```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def func_sin(x, n):
    sin_value = 0
    for i in range(n):
        coef = (-1)**i
        expo = 2*i+1 
        num = coef * x**expo
        den = math.factorial(expo)
        term = num / den
        sin_value += term
        
    return sin_value

x = np.linspace(-np.pi, np.pi, 500)

y = func_sin(x, 3)
```

```python
SMALL_SIZE = 10
MEDIUM_SIZE = 16
BIGGER_SIZE = 20

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=BIGGER_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=MEDIUM_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

```python
plt.figure(figsize=(18, 10), dpi= 150, facecolor='w', edgecolor='k')

plt.plot(x, np.sin(x), label="Sine function")

for i in range(1,6):    
    plt.plot(x, func_sin(x,i), linestyle='--', label="Taylor. Order: {}".format(i))
    plt.legend(loc="upper right")
    plt.ylim([-1.2,1.2])
    plt.title("Taylor expansion of sine function")
    
plt.savefig('Sine.png', dpi=150)

plt.show()
```
